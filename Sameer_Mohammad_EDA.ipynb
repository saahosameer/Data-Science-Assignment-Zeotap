# Step 1 : Loading the Data

```python
import pandas as pd

# Load the datasets
customers_df = pd.read_csv("Customers.csv")
products_df = pd.read_csv("Products.csv")
transactions_df = pd.read_csv("Transactions.csv")

# Display the first few rows of each dataset
print("Customers Data:")
print(customers_df.head())

print("\nProducts Data:")
print(products_df.head())

print("\nTransactions Data:")
print(transactions_df.head())
```

# Step 2: Perform Exploratory Data Analysis (EDA)
# Step 2.1: Checking for Missing Value

```python
# Check for missing values in each dataset
print("Missing values in Customers Data:")
print(customers_df.isnull().sum())

print("\nMissing values in Products Data:")
print(products_df.isnull().sum())

print("\nMissing values in Transactions Data:")
print(transactions_df.isnull().sum())
```

# Step 2.2: Basic Statistics

```python
# Basic statistics for Customers Data
print("Customers Data Statistics:")
print(customers_df.describe())

# Basic statistics for Products Data
print("\nProducts Data Statistics:")
print(products_df.describe())

# Basic statistics for Transactions Data
print("\nTransactions Data Statistics:")
print(transactions_df.describe())
```

# Step 2.3: Unique Values

```python
# Unique values in Customers Data
print("Unique Regions in Customers Data:")
print(customers_df['Region'].unique())

# Unique Categories in Products Data
print("\nUnique Categories in Products Data:")
print(products_df['Category'].unique())
```

# Step 2.4: Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")

# Plot the distribution of customer regions
plt.figure(figsize=(10, 6))
sns.countplot(data=customers_df, x='Region')
plt.title("Distribution of Customers by Region")
plt.show()

# Plot the distribution of product categories
plt.figure(figsize=(10, 6))
sns.countplot(data=products_df, x='Category')
plt.title("Distribution of Products by Category")
plt.show()

# Plot the distribution of transaction quantities
plt.figure(figsize=(10, 6))
sns.histplot(transactions_df['Quantity'], bins=30, kde=True)
plt.title("Distribution of Transaction Quantities")
plt.show()
```
